$$ \vec{f}(x,y) = \sin(x) \hat{i} + \cos(x) \hat{j}  $$

```python
import numpy as np
x_range = y_range = np.linspace(0,2*np.pi,10)
x,y = np.meshgrid(x_range,y_range)
fx,fy = np.sin(x), np.cos(y)
```

```python
from matplotlib import pyplot as plt
plt.quiver(x,y,fx,fy)
plt.axis("scaled")
plt.show()
```
