## Use Pythagoras

$$ (r+h)^2 = d^2 + r^2 $$ 

$$ \text{Radius of Earth } r = \frac{d^2 - h^2}{2h} $$

```python
h=2
d=5000
r=round((d**2 - h**2)/(2*h*1000),0)
print(f"Radius of Earth is ~ {r} km")
```

# Work out Distance to Sun
Newtons Law of Gravitation 
$$ \text{ Mass of Sun } M_s \quad \text{ Mass of Earth  } M_e \quad \text{Distance to Sun } d \text{ Force F Gravitational Constant G } $$
$$ F_g = G \frac{M_s M_e}{d^2} $$

## Centripetal force keeping Earth in orbit
$$ F_c = \frac{m v^2}{d} $$

## Equating the two forces

$$ \frac{G M_s M_e}{r^2} = \frac{M_e v^2}{r} $$
$$ \frac{G M_s}{d} = v^2 $$
$$ v = \frac{2\pi d}{T} $$


$$  d = \left( \frac{G M_s T^2}{4\pi^2} \right)^{1/3}   $$
